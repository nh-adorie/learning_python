1. import data, first look at the data

```python
catBreedDirty = pd.read_csv("/home/gitpod/.cache/kagglehub/datasets/joannanplkrk/its-raining-cats/versions/8/cat_breeds_dirty.csv", sep = ";")
catBreedDirty
```

```python
catBreedDirty = catBreedDirty.drop_duplicates() #Loại bỏ những hàng có tất cả giá trị trùng lặp
```

```python
for x in catBreedDirty.index:
    if catBreedDirty.loc[x, "Age_in_years"] < 0:
        catBreedDirty = catBreedDirty.drop(x)
# Xóa những hàng có giá trị năm tuổi của mèo là số âm (vì vô lý)
```

```python
catBreedDirty.loc[:, 'Age_in_months'] = catBreedDirty['Age_in_years'] * 12
# : ở vị trí indexer hàng có nghĩa là chọn tất cả các hàng
```

```python
catBreedDirty = catBreedDirty.dropna(subset=['Gender', 'Neutered_or_spayed'])
catBreedDirty

# Xóa những hàng có giá trị Gender và Neutered_or_spayed là NaN 
```
